**Question 3**

La signature d’un morceau est construite à partir de vecteurs de hachage de la forme (ti−ta, fa, fi), associés au temps ta de l’ancre.

On remarque que la composante temporelle du hash repose uniquement sur la différence de temps ti−ta, les fréquences fa et fi sont indépendantes de la position absolue dans le temps.

Ainsi, si l’on effectue une translation temporelle du signal (par exemple en décalant un extrait dans le temps), les valeurs ta et ti sont modifiées de la même quantité, mais leur différence ti−ta reste inchangée.

Par conséquent, les hashs (ti−ta, fa, fi) sont invariants par translation temporelle,
mais les temps associés ta ne le sont pas.

Conclusion :
La représentation est partiellement invariante par translation : les codes de hachage le sont, mais pas leur position temporelle absolue.
